```python
import os
import requests
from PIL import Image
from io import BytesIO

# Set up your Google Custom Search API key and the Search Engine ID
API_KEY = 'AIzaSyD8AGpiffQ2q3VePmogA015uWTy2WpcwEw'  # Replace with your Google API Key
CX = '57e421fc7d81947f8'  # Replace with your Custom Search Engine ID
#<script async src="https://cse.google.com/cse.js?cx=57e421fc7d81947f8">
#</script>
#<div class="gcse-search"></div>
# Function to search images via Google Custom Search API
def search_images(query, num_images=5):
    search_url ="https://www.googleapis.com/customsearch/v1?q=smoke&searchType=image&key=AIzaSyD8AGpiffQ2q3VePmogA015uWTy2WpcwEw&cx=57e421fc7d81947f8"
  #"https://www.googleapis.com/customsearch/v1"

    params = {
        'q': query,  # Search query (in this case, "smoke")
        'searchType': 'image',  # Specify that we want image results
        'key': API_KEY,  # Google API key
        'cx': CX,  # Custom search engine ID
        'num': num_images  # Number of images to retrieve
    }
    
    response = requests.get(search_url, params=params)
    search_results = response.json()
    
    # Check if the response contains items
    if 'items' in search_results:
        image_urls = [item['link'] for item in search_results['items']]
        return image_urls
    else:
        print("No images found.")
        return []

# Function to download and save images from URLs
def download_image(url, save_path):
    try:
        # Send a GET request to the image URL
        response = requests.get(url)
        
        # If the request was successful (status code 200)
        if response.status_code == 200:
            # Open the image from the response content
            image = Image.open(BytesIO(response.content))
            
            # Save the image to the specified path
            image.save(save_path)
            print(f"Image saved to {save_path}")
        else:
            print(f"Failed to retrieve image. HTTP Status Code: {response.status_code}")
    except Exception as e:
        print(f"Error occurred: {e}")

# Main function to search and download images
def download_images(query, num_images=5, save_directory="downloaded_images"):
    if not os.path.exists(save_directory):
        os.makedirs(save_directory)  # Create directory if it doesn't exist
    
    # Get image URLs from Google Custom Search
    image_urls = search_images(query, num_images)
    
    # Download and save the images
    for i, url in enumerate(image_urls):
        save_path = os.path.join(save_directory, f"image_smoke_{i+1}.jpg")
        download_image(url, save_path)

# Example usage
download_images("smoke", num_images=10, save_directory=r"C:\Users\Spacelab3\Desktop\envs\Classifier\Smokes")
```
