```python
import os

def rename_files_in_folder(folder_path, new_name_base):
    files = os.listdir(folder_path)
    for idx, file in enumerate(files):
        if idx >= 93:
            continue
        old_path = os.path.join(folder_path, file)
        file_ext = os.path.splitext(file)[1]  # Get file extension
        new_name = f"{new_name_base}_{idx}{file_ext}"
        new_path = os.path.join(folder_path, new_name)
        os.rename(old_path, new_path)
    print(f"Renamed all files in {folder_path} to pattern: {new_name_base}_X")

# Example usage
folder_path = r"C:\Users\Spacelab3\Desktop\envs\Classifier\NoSmokes"
rename_files_in_folder(folder_path, "NoSmokes_camera_2")
```

```python
import os
import shutil
output_dir = r"C:\Users\Spacelab3\Desktop\envs\MoreTraffic"

# Define the source directory (change this to your directory)
for directory in os.listdir(output_dir):
    if "sec" in directory:
        source_dir = f"C:/Users/Spacelab3/Desktop/envs/MoreTraffic/{directory}"
        # Define target folders
        text_folder = os.path.join(output_dir, "labels")
        image_folder = os.path.join(output_dir, "images")

        # Create the folders if they don't exist
        os.makedirs(text_folder, exist_ok=True)
        os.makedirs(image_folder, exist_ok=True)

        # Define image extensions
        image_extensions = (".jpg", ".jpeg", ".png", ".gif", ".bmp", ".tiff", ".webp")

        # Loop through files in the directory
        for file in os.listdir(source_dir):
            file_path = os.path.join(source_dir, file)

            # Skip directories
            if os.path.isdir(file_path):
                continue

            # Move text files
            if file.lower().endswith(".txt"):
                shutil.move(file_path, os.path.join(text_folder, file))

            # Move images
            elif file.lower().endswith(image_extensions):
                shutil.move(file_path, os.path.join(image_folder, file))

        print("Files have been sorted successfully!")
    
```
